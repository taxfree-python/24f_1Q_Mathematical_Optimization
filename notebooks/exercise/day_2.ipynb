```python
import pulp as pl
```

```python
# 問題の定義
prob = pl.LpProblem("day2_exercise", pl.LpMaximize)

# 変数の定義
x_1 = pl.LpVariable("x_1", lowBound=0)
x_2 = pl.LpVariable("x_2", lowBound=0)

# 目的関数の定義
prob += 10*x_2

# 制約条件の定義
prob += -7*x_1 + 5*x_2 <= 1
prob += -1*x_1 + x_2 <= 3

# 問題を解く
prob.solve()

# 結果の表示
print(f"{pl.LpStatus[prob.status]} solution:")
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Objective = ", pl.value(prob.objective))
```

```python
# 問題の定義
prob = pl.LpProblem("day2_exercise", pl.LpMaximize)

# 変数の定義
x_1 = pl.LpVariable("x_1", lowBound=0)
x_2 = pl.LpVariable("x_2", lowBound=0)
x_3 = pl.LpVariable("x_3", lowBound=0)
x_4 = pl.LpVariable("x_4", lowBound=0)

# 目的関数の定義
prob += x_1 - 4*x_2 - 8*x_4

# 制約条件の定義
prob += x_1 + 12*x_2 - 2*x_3 - 12*x_4 <= 0
prob += 0.25*x_1 + x_2 - 0.25*x_3 - 2*x_4 <= 0
prob += x_1 - 4*x_2 - 8*x_4 <= 1

# 問題を解く
prob.solve()

# 結果の表示
print(f"{pl.LpStatus[prob.status]} solution:")
for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Objective = ", pl.value(prob.objective))
```
